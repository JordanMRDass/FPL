```python
# Highest goals and assists adding to existing FPL_API dataset

import asyncio
import aiohttp
from colorama import Fore, init
from prettytable import PrettyTable
from fpl import FPL
import pandas as pd
import sys

df = pd.read_csv("FPL_API.csv")

async def main():
    async with aiohttp.ClientSession() as session:
        fpl = FPL(session)
        players = await fpl.get_players()

    top_performers = sorted(
        players, key=lambda x: x.goals_scored + x.assists, reverse=True)
    
    G_df = []
    A_df = []
    GA_df = []
    for player_df in list(df["Players"]):
        num = 0
        for player in top_performers:
            player_clean = str(player).split(' - ')[0]
            if player_df == player_clean and num == 0:
                G_df.append(player.goals_scored)
                A_df.append(player.assists)
                GA_df.append(player.goals_scored + player.assists)
                num += 1
    df["Goals scored"] = G_df
    df["Assists"] = A_df
    df["Goals and Assists"] = GA_df
    df.to_csv("FPL_API.csv", index = False)
    print("FPL_API updated")
    return df[["Players", "Goals scored", "Assists", "Goals and Assists"]].sort_values(by = ["Goals and Assists", "Goals scored"], ascending = [False, False])


                
        
        
await main()
```

```python
# Updating FDR

import asyncio
import aiohttp
from colorama import Fore, init
from prettytable import PrettyTable
from fpl import FPL
import pandas as pd
import sys

async def main():
    team_list = []
    all_H = []
    all_A = []
    gkp_H = []
    gkp_A = []
    def_H = []
    def_A = []
    mid_H = []
    mid_A = []
    fwd_H = []
    fwd_A = []
    async with aiohttp.ClientSession() as session:
        fpl = FPL(session)
        fdr = await fpl.FDR()
        
    for team, positions in fdr.items():
        team_list.append(team)
        all_H.append('{:.2f}'.format(float(positions['all']['H'])))
        all_A.append('{:.2f}'.format(float(positions['all']['A'])))
        gkp_H.append('{:.2f}'.format(float(positions['goalkeeper']['H'])))
        gkp_A.append('{:.2f}'.format(float(positions['goalkeeper']['A'])))
        def_H.append('{:.2f}'.format(float(positions['defender']['H'])))
        def_A.append('{:.2f}'.format(float(positions['defender']['A'])))
        mid_H.append('{:.2f}'.format(float(positions['midfielder']['H'])))
        mid_A.append('{:.2f}'.format(float(positions['midfielder']['A'])))
        fwd_H.append('{:.2f}'.format(float(positions['forward']['H'])))
        fwd_A.append('{:.2f}'.format(float(positions['forward']['A'])))
    
    FDR = pd.DataFrame({"Team": team_list, "All (Home)": all_H, "All (Away)": all_A, "Gkp (Home)": gkp_H, "Gkp (Away)": gkp_A, "Def (Home)": def_H, "Def (Away)": def_A, "Mid (Home)": mid_H, "Mid (Away)": mid_A, "Fwd (Home)": fwd_H, "Fwd (Away)": fwd_H})
    FDR.to_csv("FDR.csv")
    
if __name__ == "__main__":
    if sys.version_info >= (3, 7):
        # Python 3.7+
        await main()
    else:
        # Python 3.6
        loop = asyncio.get_event_loop()
        loop.run_until_complete(main())
        
```
