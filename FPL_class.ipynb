```python
# Updating FDR

import asyncio
import aiohttp
from colorama import Fore, init
from prettytable import PrettyTable
from fpl import FPL
import pandas as pd
import sys

async def main():
    team_list = []
    all_H = []
    all_A = []
    gkp_H = []
    gkp_A = []
    def_H = []
    def_A = []
    mid_H = []
    mid_A = []
    fwd_H = []
    fwd_A = []
    async with aiohttp.ClientSession() as session:
        fpl = FPL(session)
        fdr = await fpl.FDR()
        
    for team, positions in fdr.items():
        team_list.append(team)
        all_H.append('{:.2f}'.format(float(positions['all']['H'])))
        all_A.append('{:.2f}'.format(float(positions['all']['A'])))
        gkp_H.append('{:.2f}'.format(float(positions['goalkeeper']['H'])))
        gkp_A.append('{:.2f}'.format(float(positions['goalkeeper']['A'])))
        def_H.append('{:.2f}'.format(float(positions['defender']['H'])))
        def_A.append('{:.2f}'.format(float(positions['defender']['A'])))
        mid_H.append('{:.2f}'.format(float(positions['midfielder']['H'])))
        mid_A.append('{:.2f}'.format(float(positions['midfielder']['A'])))
        fwd_H.append('{:.2f}'.format(float(positions['forward']['H'])))
        fwd_A.append('{:.2f}'.format(float(positions['forward']['A'])))
    
    FDR = pd.DataFrame({"Team": team_list, "All (Home)": all_H, "All (Away)": all_A, "Gkp (Home)": gkp_H, "Gkp (Away)": gkp_A, "Def (Home)": def_H, "Def (Away)": def_A, "Mid (Home)": mid_H, "Mid (Away)": mid_A, "Fwd (Home)": fwd_H, "Fwd (Away)": fwd_H})
    FDR.to_csv("FDR.csv")
    
if __name__ == "__main__":
    if sys.version_info >= (3, 7):
        # Python 3.7+
        await main()
    else:
        # Python 3.6
        loop = asyncio.get_event_loop()
        loop.run_until_complete(main())
```

```python
# Updating ICT index, takes an incredibly long time, having to open 610 players online API data

import datetime
import pandas
import requests
import json
from fpl import FPL
import aiohttp
import asyncio
import re
import unicodedata
import pandas as pd

player_list = []
team_list = []
element_num_list = []
clean_name_list = []
influence_list = []
creativity_list = []
threat_list = []
ict_index_list = []

session = requests.session()
fpl = FPL(session)

def strip_accents(s):
       return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')


for num in range(1, 611):
    player_raw = await fpl.get_player(num)
    player = str(player_raw).split(' - ')[0]
    team = str(player_raw).split(' - ')[-1]
    team_list.append(team)
    player_list.append(player)
    clean_name_list.append(strip_accents(player).lower())
    element_num_list.append(num)
    url = "https://fantasy.premierleague.com/api/element-summary/"+ str(num) +"/"
    response = requests.get(url)
    data = json.loads(response.content)
    dat = data['history'][-1]
    influence_list.append(dat['influence'])
    creativity_list.append(dat['creativity'])
    threat_list.append(dat['threat'])
    ict_index_list.append(dat['ict_index'])

    
    
    
FPL_API = pd.DataFrame({"Number": element_num_list, "Players": player_list, "Clean name": clean_name_list, "Influence": influence_list, "Creativity": creativity_list, "Threat": threat_list, "ICT index": ict_index_list, "Team": team_list}).sort_values(by = ["ICT index", "Influence"], ascending = [False, False])
FPL_API.to_csv("FPL_API.csv", index = False)
print("threat, creativity, influence, ICT")

# Highest goals and assists adding to existing FPL_API dataset

import asyncio
import aiohttp
from colorama import Fore, init
from prettytable import PrettyTable
from fpl import FPL
import pandas as pd
import sys

df = pd.read_csv("FPL_API.csv")

async def main():
    async with aiohttp.ClientSession() as session:
        fpl = FPL(session)
        players = await fpl.get_players()

    top_performers = sorted(
        players, key=lambda x: x.goals_scored + x.assists, reverse=True)
    
    G_df = []
    A_df = []
    GA_df = []
    for player_df in list(df["Players"]):
        num = 0
        for player in top_performers:
            player_clean = str(player).split(' - ')[0]
            if player_df == player_clean and num == 0:
                G_df.append(player.goals_scored)
                A_df.append(player.assists)
                GA_df.append(player.goals_scored + player.assists)
                num += 1
    df["Goals scored"] = G_df
    df["Assists"] = A_df
    df["Goals and Assists"] = GA_df
    df.to_csv("FPL_API.csv", index = False)
    print("FPL_API updated")
    return df[["Players", "Goals scored", "Assists", "Goals and Assists"]].sort_values(by = ["Goals and Assists", "Goals scored"], ascending = [False, False]).head(10)


                
        
        
await main()
```
